1.  Create a dictionary from two lists: one list of keys and one list of values. Handle the case where lengths differ.

```python
l1=[1,2,3,4,5,6]
l2=['aj','bj','cj','dj','ej']
d1=dict(zip(l1,l2))
print(d1)
```

```python
from itertools import zip_longest

result = {k: v for k, v in zip_longest(l1,l2)}
print(result)
```

2. Given a sentence, build a dictionary that maps each character to its frequency.


```python
sen="Given a sentence, build a dictionary that maps each word to its frequency"
result={x:sen.count(x) for x in sen}
print(result)
```

3. Given a sentence, build a dictionary that maps each word to its frequency.


```python
sen="Given a sentence, build a dictionary that maps each word to its frequency"
words=sen.split()
result={word:words.count(word) for word in words}
print(result)
```

```python
from collections import Counter

result=dict(Counter(sen.split()))
print(result)
```

4. Given a dictionary of products and prices, increase all prices by 10%.

```python
d={'apple':70,'mango':80,'banana':50}
updated={item:price*0.1+price for item, price in d.items() }
print(updated)
```

5. Given a dictionary of products and prices, create a new dictionary with only those products whose price is above a given threshold.

```python
d={'apple':70,'mango':80,'banana':50}
updated={item:value for item,value in d.items() if value>50 }
print(updated)
```

```python
d={'apple':70,'mango':80,'banana':50}
up={}
for i,j in d.items():
    if j>50:
        up[i]=j

print(up)         
```

8. Merge two dictionaries but, if a key exists in both, sum their values.

```python
b={'cherry':200,'guava':60,'mandarin':200}
d={'apple':70,'mango':80,'banana':50,'guava':40}
b.update(d)
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}
merged = d1.copy()
for k, v in d2.items():
    merged[k] = merged.get(k, 0) + v

print(merged)    
```

9. Given a dictionary, create another dictionary where keys and values are swapped. How to handle duplicate values?
A (basic, last key wins):

```python
d = {'a': 1, 'b': 2, 'c': 1}
inverted={v:k for k,v in d.items()}
print(inverted)
```
