8. Check if two strings are anagrams.

```python
# using function
def are_anagrams(s1,s2):
    return sorted(s1)==sorted(s2)

print(are_anagrams('listen','silent'))
print(are_anagrams('hello','world'))
```

```python
from collections import Counter
def are_anagrams(s1,s2):
    s1=s1.replace(' ','').lower()
    s2=s2.replace(' ','').lower()
    return Counter(s1)==Counter(s2)

print(are_anagrams("Dormitory", "Dirty room"))    
```

11. Convert a string to an integer without using int()

```python
def str_to_int(s):
    sign=1
    i=0
    if s[0]=="-":
        sign=-1
        i=1
    elif s[0]=='+':
        i=1 

    num=0
    while i<len(s):
        digit = ord(s[i]) - ord('0')
        num = num *10+digit
        i+=1
    return sign*num 

print(str_to_int("1234")) # 1234 
print(str_to_int("-567")) # -567
```

12. Find factorial of a number (iterative and recursive)

```python
# iterative
num=5
ans=1
for i in range(1,num+1):
    ans=ans*i
    
print(ans)    
```

```python
# recursive
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*fact(n-1)
print(fact(5))    
```

13. Check if a number is prime

```python

num=int(input('Enter the number: '))
def is_prime(num):
    if num<=1:
        print('number is not prime')
        return
    for i in range (2,int(num/2)+1):
        if num%i==0:
            print('Number is not prime')
            break
    else:
        print('Number is prime')
is_prime(num)   
```

14. List all prime numbers up to N (Sieve of Eratosthenes)

```python
num=int(input('Enter upto which number you want prime numbers: '))

for i in range (2,num+1):
    is_prime=True
    for j in range(2,i):
        if i%j==0:
            is_prime=False
            break
    if is_prime:
        print(i, end=" ")        
```

17. Remove duplicates from a list while preserving order.

```python
# by using set
l1=[1,2,3,4,5,5,4,7,8,14,2,3]
print(list(set(l1)))
```

```python
def remove_duplicates(l):
   
    result=[]
    for i in l:
        if i not in result:
            result.append(i)

    return result


numbers = [1, 2, 3, 2, 4, 1, 5, 3]
print(remove_duplicates(numbers))
```

18. Count frequency of elements in a list

```python
l1=[1,2,3,4,5,5,4,7,8,14,2,3]
l2=set(l1)
freq={}
for i in l2:
    count=0
    for j in l1:
        if j==i:   
            count=count+1
    freq.update([(i ,count)])
print(freq)
```

19. Given a list, rotate it right by k positions.

```python
def rotate(lst,k):
    n=len(lst)
    k=k%n
    return lst[-k:]+lst[:-k]

lst = [1, 2, 3, 4, 5]
k = 2
print(rotate(lst, k))
```
