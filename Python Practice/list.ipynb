### 3. Largest / smallest element
Find the largest and smallest element in a list **without** using `max()` and `min()`.

```python
li=[12, 5, 7, 25, 3, 18]
lar=li[0]
sma=li[0]
for i in li:
    if i>lar:
        lar=i
    if i<sma:
        sma=i
print("Largest element:", lar)
print("Smallest element:", sma)    
```

### 5. Count all occurrences
Given a list, count how many times each element appears.

```python
li=[1,2,2,3,4,4,5,6,1,2,4,8,5,52,1,2]


for i in li:
    count=0
    for j in li:
        if j==i:
            count=count+1
    print(f"{i} : {count}")        
 
```

```python
li = [1, 2, 2, 3, 4, 4, 5, 6, 1, 2, 4, 8, 5, 52, 1, 2]

freq={}

for i in li:
    if i in freq:
        freq[i]+=1
    else:
        freq[i]=1

for key,value in freq.items():
    print(f"{key} : {value}")            
```

### 7. Reverse a list

### 8. Even / odd split
Given a list of integers, create two new lists:  
one with all even numbers and one with all odd numbers.

```python
li=[1,2,3,4,5,6,7,8,9]

even=[i for i in li if i%2==0]
odd=[i for i in li if i%2!=0]
print(even)
print(odd)
```

### 9. Second largest element
Find the second largest element in a list.

```python
li=[1,2,3,4,5,6,7,8,9]
largest=second=float('-inf')

for num in li:
    if num>largest:
        second=largest
        largest=num
    elif num>second and num!=largest:
        second=num    

print(second)
```

### 10. List comprehension practice
From a list of numbers, create:
- a list of their squares  
- a list of only even squares

```python
li=[1,2,3,4,5,6,7,8,9]

sq=[i**2 for i in li]
even_sq=[i for i in sq if i%2==0]
print(sq)
print(even_sq)
```

### 13. Flatten a nested list
Given `[[1, 2], [3, 4], [5, 6]]`, produce `[1, 2, 3, 4, 5, 6]`.

```python
li=[[1, 2], [3, 4], [5, 6]]
flatten=[j for i in li for j in i]
print(flatten)
```

```python
# using loop
flat=[]
for sub in li:
    for item in sub:
        flat.append(item)
print(flat)        
```

### 14. List rotation
Rotate a list to the right by `k` steps.  
Example: `[1, 2, 3, 4, 5]` rotated by `2` → `[4, 5, 1, 2, 3]`.

```python
from collections import deque

dq =deque([1,2,3,4,5])
dq.rotate(2)
li=list(dq)
print(li)
```

```python
li = [1,2,3,4,5]
k = 2
if li:
    k %= len(li)
    rotated = li[-k:] + li[:-k]   # returns a new list
else:
    rotated = li
print(rotated)
```

### 17. Split list into chunks
Split a list into sublists of size `n`.

Example:  
Input: `[1, 2, 3, 4, 5, 6, 7]`, `n = 3`  
Output: `[[1, 2, 3], [4, 5, 6], [7]]`

```python
li = [1, 2, 3, 4, 5, 6, 7]
n = 3

chunks=[li[i:i+n] for i in range(0,len(li),n)]
print(chunks)
```
