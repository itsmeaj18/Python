1.  Sort a dictionary by values (ascending) and return a list of (key, value) pairs.

```python
d = {'a': 1, 'b': 2, 'c': 1}
d_sorted = dict(sorted(d.items(),key=lambda x:x[1]))
print(d_sorted)
```

6. Using dictionary comprehension, filter out all items where the value is negative

```python
d = {'a': 11, 'b': 12, 'c': 14,'d':-2,'e':-4}
d_up={i:j for i, j in d.items() if j<0}
print(d_up)
```

7. Count the frequency of each value in a list using a dictionary (without Counter).

```python
lst = [1, 2, 2, 3, 1]
freq = {}

for i in lst:
    cnt = 0               # reset counter for each i
    for j in lst:
        if j == i:
            cnt += 1      # increment correctly
    freq[i] = cnt

print(freq)
```

```python
freq={}
for x in lst:
    freq[x]=freq.get(x,0)+1
print(freq)    
```

8. Given two lists: one of employee IDs and one of departments, build a dictionary mapping each department to a list of employee IDs.

```python
emp_ids = [1, 2, 3, 4]
depts = ['IT', 'HR', 'IT', 'Finance']

dic=dict(zip(emp_ids,depts))
print(dic)
```

```python
emp_ids = [1, 2, 3, 4]
depts = ['IT', 'HR', 'IT', 'Finance']
dept_map = {}
for emp, dept in zip(emp_ids, depts):
    dept_map.setdefault(dept, []).append(emp)
print(dept_map)    
```

9.  Flatten a nested dictionary one level deep.

```python
nested = {'a': {'x': 1, 'y': 2}, 'b': {'z': 3}}
flat={}
for i,j in nested.items():
    for v,k in j.items():
        flat[f"{i}_{v}"]=k
print(flat)
```

10. Given a dictionary, write code to check if all values are unique.

```python
d = {'a': 1, 'b': 2, 'c': 1,'e':5}
values=list(d.values())
all_unique=len(values)==len(set(values))
print(all_unique)
```

13. Given a list of dictionaries representing employees (id, name, dept), build a dictionary keyed by id

```python
employees = [
{'id': 1, 'name': 'Alice', 'dept': 'IT'},
{'id': 2, 'name': 'Bob', 'dept': 'HR'}
]

dic={emp['id']:emp for emp in employees}
print(dic)
```

14. Given transaction records as a list of tuples (customer_id, amount), build a dictionary mapping customer_id to total amount.

```python
transactions = [(1, 100), (2, 50), (1, 200)]
totals={}
for i, j in transactions:
    totals[i]=totals.get(i,0)+j

print(totals)    
```
