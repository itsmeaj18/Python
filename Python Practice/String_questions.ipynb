### 2. Write a program to print First non-repeated character from given String? 

```python
# Method 1
st= "hherlelno World"
for i in st:
    if st.count(i)==1:
        print("First non repeated character:",i)
        break
    

```

```python
## using Dictionary
freq={}
for i in st:
    freq[i] = freq.get(i,0)+1

for i in st:
    if freq[i]==1:
        print("First non repeated character:",i)
        break 
```

### 4. Write a program to remove the duplicate characters from the given String? 

```python
# Method 1
st= "hherlelno World"
l=[]
for i in st:
    if i not in l :
        l.append(i)
print("".join(l))
```

```python
# Method 2 - using set for fast process
seen=set()
result=[]

for ch in st:
    if ch not in seen:
        seen.add(ch)
        result.append(ch)

print("".join(result))
```

```python
# Method 3 using dictionary
print("".join(dict.fromkeys(st)))
```

### 7. write a program to print all duplicate character and their count form the given String? 

```python
# method 1
string = "programming"
count={}

for char in string:
    if char in count:
        count[char]+=1
    else:
        count[char]=1

print("Duplicate characters are: ")
for char, freq in count.items():
    if freq > 1:
        print(f"{char} -> {freq}")        
```

```python
# method 2
from collections import Counter
string = "programming"

count=Counter(string)

print("Duplicate characters are:")
for char, freq in count.items():
    if freq > 1:
        print(f"{char} -> {freq}")
```

### 8. Write a program to print Highest occurred character from given String? 

```python
from collections import Counter

string = "proggramming"
count=Counter(string)
max_char =max(count, key=count.get)
print(f"Highest occurred character: '{max_char}' → {count[max_char]} times")
```

```python
# method 2
from collections import Counter

string = "successes"
count = Counter(string)

# Find the maximum frequency
max_freq = max(count.values())

# Get all characters with that frequency
max_chars = [char for char, freq in count.items() if freq == max_freq]

print("Highest occurred character(s):", ", ".join(max_chars))
print("Frequency:", max_freq)
```
