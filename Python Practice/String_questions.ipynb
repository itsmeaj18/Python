### 1. Write a program to print the reverse of the String? 

```python
st = "Hello world"
print("method 1:",st[::-1])
print("method 2:","".join(reversed(st)))
```

```python
# using for loop
l=[]
for i in range(len(st)-1,-1,-1):
   l.append(st[i])
print("".join(l))    
```

### 2. Write a program to print First non-repeated character from given String? 

```python
# Method 1
st= "hherlelno World"
for i in st:
    if st.count(i)==1:
        print("First non repeated character:",i)
        break
    

```

```python
## using Dictionary
freq={}
for i in st:
    freq[i] = freq.get(i,0)+1

for i in st:
    if freq[i]==1:
        print("First non repeated character:",i)
        break 
```

### 3. Write a program to print last non-repeated character from given String? 

```python
# method 1
st= "hherlelno World"
reve="".join(reversed(st))
for i in reve:
    if reve.count(i)==1:
        print("Last non repeated character:",i) 
        break

```

```python
# method 2
freq={}
reve="".join(reversed(st))
for i in reve:
    freq[i]=freq.get(i,0)+1

for i in reve:
    if freq[i]==1:
        print("First non repeated character:",i)
        break     
```

### 4. Write a program to remove the duplicate characters from the given String? 

```python
# Method 1
st= "hherlelno World"
l=[]
for i in st:
    if i not in l :
        l.append(i)
print("".join(l))
```

```python
# Method 2 - using set for fast process
seen=set()
result=[]

for ch in st:
    if ch not in seen:
        seen.add(ch)
        result.append(ch)

print("".join(result))
```

```python
# Method 3 using dictionary
print("".join(dict.fromkeys(st)))
```
