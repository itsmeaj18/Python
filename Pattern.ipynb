## <b>Right-Angled Triangle

```python
n=5
```

```python
n=int(input("Enter number of rows"))
for i in range (n+1):
    print("*"*i)
```

## <b>Inverted Right-Angled Triangle

```python
n=int(input("Enter number of rows :"))
for i in range(n,0,-1):
    print("*"*i)
```

## <b>Pyramid Pattern

```python
n=int(input("Enter number of rows :"))
for i in range(n):
    print(" "*(n-i)+'*'*(i+1)+"*"*i )
    
```

## <b>Inverse Pyramid Pattern

```python
n=int(input("Enter number of rows :"))
for i in range(n):
    print(" "* i +("*"*(n-i))+"*"*(n-(i+1)))
    
```

## <b>Diamond Pattern

```python
for i in range(n):
    print(" "*(n-i)+'*'*(i+1)+"*"*i )
for j in range(1,n):
    print(" "* (j+1) +("*"*(n-j))+"*"*(n-(j+1)))
```

## <b>Hollow Square Pattern


```python
for i in range (n):
    for j in range (n):
        if j==0 or j==n-1 or i==n-1 or i==0 :
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print( )        
```
